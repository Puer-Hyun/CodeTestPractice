6660, 6661, 6662도 존재한다는 것을 인지해야한다.

가장 먼저 든 생각은 1개씩 키워가면서 연속된 3개의 6이 등장하는가-를 체크해서 result에 append 시켜주는 것이다. 바로 떠오르지는 않았다.

````python
def Shome(x : int) -> bool :
    """
    x가 666이 연속으로 들어가 있는 숫자인지를 반환합니다.
    
    Args : 
        x (int) : 인자, 단 x>=666에 대해서 실행한다. 제일 작은 종말의 수가 666이기 때문이다.
        
    Returns : bool
    
    """
    x = list(str(x))
    n = len(x)
    result = False
    # 한 번이라도 연속으로 666이 등장하면 result = True로 바꾸고 break 한다.
    for i in range(n-2):
        if x[i]==x[i+1]==x[i+2]=='6':
            result = True
            break
        else:
            pass
    return result
    
    

import time
start = time.time()
#행동
k = 666
n = 9999
answer = []
while len(answer)!=n:
    if Shome(k) :
        answer.append(k)
    k += 1
print(answer[-1])
print("spent time : {}".format(time.time()-start))
    
````
