* 각 행의 양 옆에 0 이라는 padding을 넣어둔 뒤에, data와 같은 사이즈를 갖는 result라는 데이터를 만들어서, result[i][j] = data[i][j]+max(result[i-1][j-1], result[i-1][j])를 시키자.

```python
#백준의 예제 입력을 text로 받아서 한 번에 data로 뽑아내고 싶기 때문
given_data ='''5
7
3 8
8 1 0
2 7 4 4
4 5 2 6 5'''

# mode가 False면 기본적으로 위에 given_data를 백준에서 가져와서 쓸거고
# 그 말은 given_data가 주어져있다는 것이고
# 백준에 옮겨쓸때는 n,data = solution(True)로 시작해야함.
# num은 기본적으로 1로 설정해두었으나, num의 의미는 1행에 들어가는 숫자의 개수

def data_preprocessing(row_num : int, data : str = None , mode : bool = True, num : int = 1) -> tuple : 
    if mode == False:
        data = given_data
        row = data.split('\n')
        if num == 1:
            parameters = int(row[0])
        else:
            parameters = list(map(int, row[0].split()))
        result = [list(map(int,i.split())) for i in row[1:]]
    else:
        if num == 1:
            parameters = int(input())
        else:
            parameters = list(map(int, input().split()))
        result = []
        for _ in range(row_num):
            result.append(list(map(int, input().split()))) 
    return parameters, result
        
            
    
data_preprocessing(5, mode = False)
```

```python
n, data = data_preprocessing(5, mode = False)
```

```python
n, data = data_preprocessing(5, mode = False)
result = []
for index in range(n):
    data[index].insert(0,0)
    data[index].append(0)
    result.append([0 for _ in range(index+3)])

for i in range(n):
    if i==0:
        result[i][1] = data[i][1]
    else:
        for j in range(i+2):
            result[i][j] = data[i][j] + max(result[i-1][j-1], result[i-1][j])
            
print(max(result[-1]))
```
