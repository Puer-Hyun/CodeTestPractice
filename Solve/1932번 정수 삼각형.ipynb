```python
#   text    data   
given_data ='''5
7
3 8
8 1 0
2 7 4 4
4 5 2 6 5'''

# mode False   given_data   
#   given_data  
#   n,data = solution(True) .
# num  1 , num  1   

def data_preprocessing(row_num : int, data : str = None , mode : bool = True, num : int = 1) -> tuple : 
    if mode == False:
        data = given_data
        row = data.split('\n')
        if num == 1:
            parameters = int(row[0])
        else:
            parameters = list(map(int, row[0].split()))
        result = [list(map(int,i.split())) for i in row[1:]]
    else:
        if num == 1:
            parameters = int(input())
        else:
            parameters = list(map(int, input().split()))
        result = []
        for _ in range(row_num):
            result.append(list(map(int, input().split()))) 
    return parameters, result
        
            
    
data_preprocessing(5, mode = False)
```
