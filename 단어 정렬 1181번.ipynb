```python
from collections import defaultdict
_input = '''13
but
i
wont
hesitate
no
more
no
more
it
cannot
wait
im
yours'''
n, data = _input.split('\n')[0], list(set(_input.split('\n')[1:]))
default_dict = defaultdict(list)
for i in data:
    default_dict[len(i)].append(i)

sorted_dict = defaultdict(list)
for k in sorted(default_dict.keys()):
    sorted_dict[k] = sorted(default_dict[k])
    
sorted_dict
for v in sorted_dict.values():
    for idx in range(len(v)):
        print(v[idx])
```

```python
#   .
from collections import defaultdict
import sys
input = sys.stdin.readline

n = int(input().strip())
data = [input().strip() for _ in range(n)]

default_dict = defaultdict(list)
for i in data:
    default_dict[len(i)].append(i)

sorted_dict = defaultdict(list)
for k in sorted(default_dict.keys()):
    sorted_dict[k] = sorted(default_dict[k])
    
for v in sorted_dict.values():
    for idx in range(len(v)):
        print(v[idx])
```
