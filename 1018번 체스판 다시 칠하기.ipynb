```python
input_ = '''11 12
BWWBWWBWWBWW
BWWBWBBWWBWW
WBWWBWBBWWBW
BWWBWBBWWBWW
WBWWBWBBWWBW
BWWBWBBWWBWW
WBWWBWBBWWBW
BWWBWBWWWBWW
WBWWBWBBWWBW
BWWBWBBWWBWW
WBWWBWBBWWBW'''
N, M = map(int,input_.split('\n')[0].split())
data = input_.split('\n')[1:]
```

```python
tmp_list =['BWBWBWBW', 'WBWBWBWB']
```

```python
def count_different_chars(s1, s2):
    count = 0
    for c1, c2 in zip(s1, s2):
        if c1 != c2:
            count += 1
    return count


def startwithB(graph, mode : bool) -> int:
    if mode: #(1,1) B 
        cnt, cnt2 = 0, 0
        for i in range(8):
            cnt += count_different_chars(tmp_list[i%2], graph[i])
            cnt2 += count_different_chars(tmp_list[(i+1)%2], graph[i])
        return min(cnt,cnt2)
    else:
        cnt, cnt2 = 0, 0
        for i in range(8):
            cnt += count_different_chars(tmp_list[i%2], graph[i])
            cnt2 += count_different_chars(tmp_list[(i+1)%2], graph[i])
        return min(cnt,cnt2)

```

```python
answer = int(1e9)
for r in range(N-7):
    for c in range(M-7):
        tmp_graph = [row[c:c+8] for row in data[r:r+8]]
        answer = min(answer, startwithB(tmp_graph, mode=(tmp_graph[0][0]=='B')))
print(answer)
```
