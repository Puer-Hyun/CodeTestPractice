```python
example = '''8
4
3
6
8
7
5
2
1'''.split('\n')
n = int(example[0])
data = list(map(int, example[1:]))

import copy
a = copy.deepcopy(data)
b = []
tmp = []
result = []

b.append(a.pop(-1))
result.append('+')

while a:
    if a[-1]>b[-1]:
        b.append(a.pop(-1))
        result.append('+')
    else:
        while b and a[-1] < b[-1]:
            tmp.append(b.pop(-1))
            result.append('-')

c = b + tmp[::-1]
if c == list(range(1,n+1)):
    result = ['-' for _ in range(len(tmp))]+result[::-1]
else:
    result = "NO"
    
print(result)
        
```
