투포인터를 사용하는 것 같은 느낌이 든다.

````python
example = '''4 4
802
743
457
539'''

def solution(num:int, data : list) -> int:
    """
    data의 각 요소를 num으로 나눈 몫들의 합을 반환한다.
    
    """
    result = 0
    for d in data:
        result += d//num
    return result

n, target = map(int,example.split('\n')[0].split())
data = list(map(int, [i for i in example.split('\n')[1:]]))

start, end = 0, max(data)

while start <= end:
    mid = (start + end) // 2
    
    if solution(mid, data) < target:
        end = mid 
        
    else:
        start = mid + 1

print(end-1)

````

# 최종 코드
```python
import sys
K, N = map(int, input().split())
data = [int(sys.stdin.readline()) for _ in range(K)]
start, end = 1, max(data)
while start <= end:
    mid = (start+end)//2
    lines = 0
    for i in data:
        lines += i // mid
    if lines >= N:
        start = mid+1
    else:
        end = mid-1
print(end)
```
