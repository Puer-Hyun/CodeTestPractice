```python
def data_preprocessing(data : str) -> tuple :
    data = given_data.split('\n')
    first_line = data[0]
    remain = data[1:]
    return first_line, remain
    
given_data = '''10
5
2
3
1
4
2
3
5
1
7'''

n, data = data_preprocessing(given_data)
n = int(n)
data = list(map(int, data))
dp = [0 for _ in range(10001)]
for i in data:
    dp[i] += 1
for i, v in enumerate(dp):
    if v==0:
        continue
    else:
        for _ in range(v):
            print(i)    
```

```python
# 여기서부터 실제 코드
import sys
input = sys.stdin.readline
n = int(input())
dp = [0 for _ in range(10001)]
data = [int(input()) for _ in range(n)]
for i in data:
    dp[i] += 1
    
for i, v in enumerate(dp):
    if v==0:
        continue
    else:
        for _ in range(v):
            print(i) 
```
